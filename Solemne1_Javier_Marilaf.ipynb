El comando columns nos muetra el encabezado de cada columna de nuestro dataset. En este caso son 5 que representan el largo y ancho tanto del sépalo como el pétalo incluida la especie de cada flor iris.

```python
#Utilice mean, mode and median

#Media
print("Promedio: ")
print("Largo del sépalo:", data["sepal.length"].mean())
print("Ancho del sépalo:", data["sepal.width"].mean())
print("Largo del pétalo:", data["petal.length"].mean())
print("Ancho del pétalo:", data["petal.width"].mean())


#Moda
print("\nModa: ")
print("Largo del sépalo:", data["sepal.length"].mode())
print("Ancho del sépalo:", data["sepal.width"].mode())
print("Largo del pétalo:", data["petal.length"].mode())
print("Ancho del pétalo:", data["petal.width"].mode())

#Mediana
print("\nMediana: ")
print("Largo del sépalo:", data["sepal.length"].median())
print("Ancho del sépalo:", data["sepal.width"].median())
print("Largo del pétalo:", data["petal.length"].median())
print("Ancho del pétalo:", data["petal.width"].median())
```

Añadir qué observaciones hay respecto del comando utilizado

El comando mean() nos muestra el promedio de una columan, en este caso se usa para cada medida. A su vez el comando mode() nos muestra la moda de cada columna y por último la instrucción median nos indica la mediana de cada conjunto de datos de una columna, muy similar a las anteriores instrucciones.

##Gráficos
Plantear algunos gráficos que le ayuden a explorar de mejor manera y genérelos con su respectivo análisis

```python
# Gráfico de densidad Largo de sépalos
data.plot(kind='scatter', x='sepal.length', y='sepal.width');
plt.show()

sns.set_style("whitegrid");
sns.FacetGrid(data, hue="variety", height=4) \
    .map(plt.scatter, "sepal.length", "sepal.width") \
    .add_legend();
plt.show();
```

Se puede apreciar que la flor iris Setosa tiene una tendencia a tener el sépalo mas ancho que las demás especies, mientras que la variedad iris Virginica tiene en promedio el sépalo mas largo. Tambien la iris Versicolor se mantiene con medidas ma pequeñas que sus contrapartes.

```python
#Gráfico para el largo y ancho de los pétalos

data.plot(kind='scatter', x='petal.length', y='petal.width');
plt.show()

sns.set_style("whitegrid");
sns.FacetGrid(data, hue="variety", height=4) \
    .map(plt.scatter, "petal.length", "petal.width") \
    .add_legend();
plt.show();
```

Segun el gráfico que representa las medidas del largo y ancho de los pétalos, se puede apreciar claramente que la flor iris Setosa,representada por el color azul, posee los pétalos mas pequeños tanto en largo como ancho. Por otra parte la iris Virginica, con un color verde, por el contrario tiene los pétalos mas grandes y anchos con diferencia. Y por último la flor iris Versicolor se encuentra en medio de las demas variedades, como muestra el grafico representada por el color anaranjado.
